```python
import os

import wfdb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import ast
```

```python
dataset_loc = "../ptb-xl"
```

```python
# !ls {dataset_loc}
```

```python
path = os.path.join(dataset_loc, "ptbxl_database.csv")
df = pd.read_csv(path, index_col='ecg_id')
df
```

```python
df.scp_codes = df.scp_codes.apply(lambda x: ast.literal_eval(x))
df.scp_codes
```

```python
agg_df = pd.read_csv(os.path.join(dataset_loc, 'scp_statements.csv'), index_col=0)
agg_df = agg_df[agg_df.diagnostic == 1]
agg_df
```

```python
def aggregate_diagnostic(y_dic):
    tmp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            tmp.append(agg_df.loc[key].diagnostic_class)
    return list(set(tmp))

# Apply diagnostic superclass
df['diagnostic_superclass'] = df.scp_codes.apply(aggregate_diagnostic)
```

```python
df['diagnostic_superclass'].value_counts()
```

```python
df = df[df['diagnostic_superclass'].apply(lambda x: x != [])]
df
```

```python
Z = pd.DataFrame(0, index=df.index, columns=['NORM', 'MI', 'STTC', 'CD', 'HYP'], dtype='int')
Z
```

```python
for i in Z.index:
    for k in df.loc[i].diagnostic_superclass:
        Z.loc[i, k] = 1
```
